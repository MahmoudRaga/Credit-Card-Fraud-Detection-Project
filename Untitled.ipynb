# "Importing Libraries"

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
plt.style.use('dark_background')
```

# "Reading Data"

```python
application_record = pd.read_csv('application_record.csv')
application_record
```

```python
credit_record = pd.read_csv('credit_record.csv')
credit_record
```

# "Drop Column (OCCUPATION_TYPE) From (Application_Record)"

```python
application_record.drop(['OCCUPATION_TYPE'],axis = 1,inplace = True)
```

# "Encoding object data with Label Encoder For (Application_Record) "

```python
LE = LabelEncoder()
application_record['CODE_GENDER'] = LE.fit_transform(application_record['CODE_GENDER'])
application_record['FLAG_OWN_CAR'] = LE.fit_transform(application_record['FLAG_OWN_CAR'])
application_record['FLAG_OWN_REALTY'] = LE.fit_transform(application_record['FLAG_OWN_REALTY'])
application_record['NAME_INCOME_TYPE'] = LE.fit_transform(application_record['NAME_INCOME_TYPE'])
application_record['NAME_EDUCATION_TYPE'] = LE.fit_transform(application_record['NAME_EDUCATION_TYPE'])
application_record['NAME_FAMILY_STATUS'] = LE.fit_transform(application_record['NAME_FAMILY_STATUS'])
application_record['NAME_HOUSING_TYPE'] = LE.fit_transform(application_record['NAME_HOUSING_TYPE'])
```

```python
application_record
```

# "Data Splitting From (Application_Record) To (X_1  & Y_1)"

```python
x_1 = application_record.iloc[:,:-1].values
x_1
```

```python
y_1 = application_record.iloc[:,-1].values
y_1
```

# " Scaling Data By Standard Scaler in X_1 "

```python
SC = StandardScaler()
x_1[:,0:1] = SC.fit_transform(x_1[:,0:1])
x_1[:,4:12] = SC.fit_transform(x_1[:,4:12])
x_1[:,15:] = SC.fit_transform(x_1[:,15:])
```

```python
df = pd.DataFrame(x_1 , columns=['ID','CODE_GENDER','FLAG_OWN_CAR','FLAG_OWN_REALTY','CNT_CHILDREN','AMT_INCOME_TOTAL','NAME_INCOME_TYPE','NAME_EDUCATION_TYPE','NAME_FAMILY_STATUS','NAME_HOUSING_TYPE','DAYS_BIRTH','DAYS_EMPLOYED','FLAG_MOBIL','FLAG_WORK_PHONE','FLAG_PHONE','FLAG_EMAIL'])
df
```

# "Data Splitting From (Application_Record) To (Training & Testing)"

```python
x_1_train,x_1_test,y_1_train,y_1_test = train_test_split(x_1 , y_1 ,test_size = 0.3 ,random_state = 0)
```

# "Predicting Data"

```python
svc = SVC()
svc.fit(x_1_train,y_1_train)
y_1_pred = svc.predict(x_1_test)
y_1_pred
```

# "Classification Report"

```python
print(classification_report(y_1_test,y_1_pred))
```

# "Compare between Actual Data & Predicted Data"

```python
df_1_comp = pd.DataFrame({'Actual':y_1_test , 'Predict':y_1_pred})
df_1_comp
```

```python
plt.figure(figsize=(10, 7))
plt.title (' Actual & Predict ',color = 'w')
sns.heatmap (df_1_comp)
sns.pairplot(df_1_comp)
```
